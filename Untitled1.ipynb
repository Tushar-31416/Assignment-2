```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

def fy(t,y):
  return t*np.exp(3*t)-2*y


t_min = 0
t_max = 1
t = [t_min]
N = 100
for i in range(N):
  t.append(t[i]+1.0/N)
print(t)


s = solve_ivp(fy,[t[0],t[-1]],[0],method='RK45',t_eval=t)


def f1(t):
    y1=((1/25)*np.exp(-2*t)*(1-np.exp(5*t)+5*np.exp(5*t)*t))
    return y1


plt.plot(t,np.vectorize(f1)(t),color='blue',label='Exact')
plt.plot(s.t,s.y[0],'--',color='red',label='Numerical')
plt.xlabel('t')
plt.ylabel('y')
plt.show()
```

```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

def fy2(t,y):
  return 1-(t-y)**2

t_min = 2
t_max = 3
t = [t_min]
N = 100
for i in range(N):
  t.append(t[i]+1.0/N)



s = solve_ivp(fy2,[t[0],t[-1]],[1],method='RK45',t_eval=t)


def f2(t):
    y2=((1-3*t+t**2)/(t-3))
    return y2


plt.plot(t,np.vectorize(f2)(t),color='blue',label='Exact')

plt.plot(s.t,s.y[0],'--',color='red',label='Numerical')
plt.xlabel('t')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

def fy(t,y):
  return 1+y/t

t_min = 1
t_max = 2
t = [t_min]
N = 100
for i in range(N):
  t.append(t[i]+1.0/N)


s = solve_ivp(fy,[t[0],t[-1]],[2],method='RK45',t_eval=t)

def f3(t):
    y3=(2*t+t*np.log(t))
    return y3


plt.plot(t,np.vectorize(f3)(t),color='blue',label='Exact')

plt.plot(s.t,s.y[0],'--',color='red',label='Numerical')
plt.xlabel('t')
plt.ylabel('y')
plt.legend()

plt.show()
```
